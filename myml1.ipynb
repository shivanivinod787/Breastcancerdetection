```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split


breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
dataframe = pd.DataFrame(breast_cancer_dataset.data,columns= breast_cancer_dataset.feature_names)
dataframe.head()
```

```python
dataframe['label']=breast_cancer_dataset.target
dataframe.tail()
```

```python
X = dataframe.drop(columns = 'label',axis =1)
Y = dataframe['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')

                            ,
])
```

```python
model.compile(optimizer= 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model_accuracy')
plt.xlabel('epochs')
plt.ylabel('Accuracy')

plt.legend(['training_data','validation_data'],loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("model_loss")
plt.xlabel('epochs')
plt.ylabel('loss')

plt.legend(['training_data','validation_data'],loc ='upper right')
```

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```
